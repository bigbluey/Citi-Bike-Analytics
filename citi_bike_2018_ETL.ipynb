# City Bike Data 2018

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Files to Load
march2018 = "201803-citibike-tripdata.csv"
june2018 = "201806-citibike-tripdata.csv"
september2018 = "201809-citibike-tripdata.csv"
december2018 = "201812-citibike-tripdata.csv"
```

```python
# Read April 2018 Citi Bike Data File & Store into Pandas DataFrame
march2018_df = pd.read_csv(march2018, encoding="ISO-8859-1")
march2018_df.head()
```

```python
march2018_df = march2018_df.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
march2018_df.head()
```

```python
# Read June 2018 Citi Bike Data File & Store into Pandas DataFrame
june2018_df = pd.read_csv(june2018, encoding="ISO-8859-1")
june2018_df.head()
```

```python
june2018_df = june2018_df.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
june2018_df.head()
```

```python
# Read September 2018 Citi Bike Data File & Store into Pandas DataFrame
september2018_df = pd.read_csv(september2018, encoding="ISO-8859-1")
september2018_df.head()
```

```python
september2018_df = september2018_df.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
september2018_df.head()
```

```python
# Read December 2018 Citi Bike Data File & Store into Pandas DataFrame
december2018_df = pd.read_csv(december2018, encoding="ISO-8859-1")
december2018_df.head()
```

```python
december2018_df = december2018_df.rename(columns={"tripduration": "Trip Duration",
                                            "starttime": "Start Time",
                                            "stoptime": "Stop Time",
                                            "start station id": "Start Station ID",
                                            "start station name": "Start Station Name",
                                            "start station latitude": "Start Station Latitude",
                                            "start station longitude": "Start Station Longitude",
                                            "end station id": "End Station ID",
                                            "end station name": "End Station Name",
                                            "end station latitude": "End Station Latitude",
                                            "end station longitude": "End Station Longitude",
                                            "bikeid": "Bike ID",
                                            "usertype": "User Type",
                                            "birth year": "Birth Year",
                                            "gender": "Gender"
                                           })
december2018_df.head()
```

```python
# Combine All 2018 DataFrames
combined2018_df = march2018_df.append([june2018_df, september2018_df, december2018_df], ignore_index=True)
combined2018_df
```
